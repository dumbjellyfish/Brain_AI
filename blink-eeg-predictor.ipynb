```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session



#This creates the path to call my EEG data

A = '/kaggle/input/blink-eeg/Averaged_Blink_Data.csv'
#Averaged data set

C = '/kaggle/input/blink-eeg/Complete_Blink_Data.csv'
#Complete data set

T = '/kaggle/input/blink-eeg/thing.csv'


Averaged = pd.read_csv(A)
Complete = pd.read_csv(C)
thing = pd.read_csv(T)
```

```python
#Averaged
#Complete
#This simply prints my data
thing
```

```python

yA = Averaged.Blink_Averaged

yC = Complete.Blink
#This assigns my output/y value, to the blink values in my data set.
# (this is important for the testing and training data)

featuresA = ['Average_1','Average_0','Average_3']
featuresC = ['Channel_0','Channel_1','Channel_3']


XA = Averaged[featuresA]
XC = Complete[featuresC]
#This assigns the EEG chanels 0,1 & 3 as the input values/x value




#derivative
yT = thing.Blink_Averaged
featuresT = ['0_Der_avg', '1_Der_avg','3_Der_avg']
F1 = ['0_Der_avg', '1_Der_avg','3_Der_avg']
XT = thing[F1]
```

```python
from sklearn.model_selection import train_test_split


train_XA, val_XA, train_yA, val_yA = train_test_split(XA,yA, random_state=2)

train_XC, val_XC, train_yC, val_yC = train_test_split(XC,yC, random_state=2)

train_XT, val_XT, train_yT, val_yT = train_test_split(XT,yT, random_state=2)
#This splits my data into training and testing data. The model will adjust
#it's weights based on the training data and obtain it's score based on the 
#testing data
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
#for this project, I am using a decision tree c. I might also want
#to try a random forest as this can be a more accurate model

EEG_modelA = DecisionTreeClassifier(random_state=10)
EEG_modelC = DecisionTreeClassifier(random_state=10)

EEG_modelT = RandomForestClassifier(random_state=20)


EEG_modelA.fit(train_XA,train_yA)
EEG_modelC.fit(train_XC,train_yC)

EEG_modelT.fit(train_XT,train_yT)


#now, our model is being fit on our training data
```

```python
from sklearn.metrics import classification_report


y_predT = EEG_modelT.predict(val_XT)
reportT = classification_report(val_yT, y_predT)
print('Classification report for Random Forest', '\n'*2, reportT, '\n')

```

```python
from sklearn.metrics import classification_report

print("This is The Classification Report for the Decision Tree", '\n'*3)

y_predA = EEG_modelA.predict(val_XA)
reportA = classification_report(val_yA, y_predA)
print('Classification report for Averaged Data', '\n'*2, reportA, '\n')


y_predC = EEG_modelC.predict(val_XC)
reportC = classification_report(val_yC,y_predC)
print('Classification report for Complete Data','\n'*2,reportC)
```
